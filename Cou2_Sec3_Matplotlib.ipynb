```python
import numpy as np
import matplotlib.pyplot as plt
```

---
### 12 - Line Chart
---

```python
x = np.linspace(0,20,1000)
```

```python
y = np.sin(x)+0.2*x
```

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('myplot');
```

---
### 13 - Scatterplot
---

```python
x = np.random.randn(100,2)
x[:50] += 3
plt.scatter(x[:,0], x[:,1]);
```

```python
x = np.random.randn(100,2)
x[:50] += 3
y = np.zeros(100)
y[:50] = 1
plt.scatter(x[:,0], x[:,1], c=y);
```
